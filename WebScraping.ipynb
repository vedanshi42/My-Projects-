```python
#Python program to scrape website
#and save quotes from website
import requests
from bs4 import BeautifulSoup
import csv

URL = "https://earthdata.nasa.gov/earth-observation-data/near-real-time/hazards-and-disasters"
r = requests.get(URL)

soup = BeautifulSoup(r.content, 'html5lib')

hazards_and_disasters=[] # a list to store quotes

table = soup.find('table')

for row in table.findAll('tr'):
	hazard = {}
	
	hazard['url'] = row.a['href']
	hazard['img'] = row.img['src']
	hazard['lines'] = row.img['alt'].split(" #")[0]
	hazards_and_disasters.append(hazard)

filename = 'hazards and disasters.csv'
with open(filename, 'w', newline='') as f:
	w = csv.DictWriter(f,['url','img','lines'])
	w.writeheader()
	for hazard in hazards_and_disasters:
		w.writerow(hazard)
```
